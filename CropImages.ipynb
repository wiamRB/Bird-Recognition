```python
import zipfile
import os

def zip_folder(folder_path, output_path):
    """
    Zips the contents of a folder to a zip file
    """
    with zipfile.ZipFile(output_path, 'w', zipfile.ZIP_DEFLATED) as zip_file:
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                file_path = os.path.join(root, file)
                zip_file.write(file_path, arcname=os.path.relpath(file_path, folder_path))
```

```python
from typing import List

import numpy as np


# converts Detections into format that can be consumed by match_detections_with_tracks function
def detections2boxes(detections: Detections) -> np.ndarray:
    return np.hstack((
        detections.xyxy,
        detections.confidence[:, np.newaxis]
    ))


# converts List[STrack] into format that can be consumed by match_detections_with_tracks function
def tracks2boxes(tracks: List[STrack]) -> np.ndarray:
    return np.array([
        track.tlbr
        for track
        in tracks
    ], dtype=float)


# matches our bounding boxes with predictions
def match_detections_with_tracks(
    detections: Detections,
    tracks: List[STrack]
) -> Detections:
    if not np.any(detections.xyxy) or len(tracks) == 0:
        return np.empty((0,))

    tracks_boxes = tracks2boxes(tracks=tracks)
    iou = box_iou_batch(tracks_boxes, detections.xyxy)
    track2detection = np.argmax(iou, axis=1)

    tracker_ids = [None] * len(detections)

    for tracker_index, detection_index in enumerate(track2detection):
        if iou[tracker_index, detection_index] != 0:
            tracker_ids[detection_index] = tracks[tracker_index].track_id

    return tracker_ids
```

```python
from typing import List, Optional, Union

import cv2
import numpy as np

from supervision.draw.color import Color, ColorPalette


class BoxAnnotator:
    def __init__(
        self,
        color: Union[Color, ColorPalette],
        thickness: int = 2,
        text_color: Color = Color.black(),
        text_scale: float = 0.5,
        text_thickness: int = 1,
        text_padding: int = 10,
    ):
        """
        A class for drawing bounding boxes on an image using detections provided.
        :param color: Union[Color, ColorPalette] :  The color to draw the bounding box, can be a single color or a color palette
        :param thickness: int :  The thickness of the bounding box lines, default is 2
        :param text_color: Color :  The color of the text on the bounding box, default is white
        :param text_scale: float :  The scale of the text on the bounding box, default is 0.5
        :param text_thickness: int :  The thickness of the text on the bounding box, default is 1
        :param text_padding: int :  The padding around the text on the bounding box, default is 5
        """
        self.color: Union[Color, ColorPalette] = color
        self.thickness: int = thickness
        self.text_color: Color = text_color
        self.text_scale: float = text_scale
        self.text_thickness: int = text_thickness
        self.text_padding: int = text_padding

    def annotate(
        self,
        frame: np.ndarray,
        detections: Detections,
        labels: Optional[List[str]] = None,
    ) -> np.ndarray:
        """
        Draws bounding boxes on the frame using the detections provided.
        :param frame: np.ndarray : The image on which the bounding boxes will be drawn
        :param detections: Detections : The detections for which the bounding boxes will be drawn
        :param labels: Optional[List[str]] :  An optional list of labels corresponding to each detection. If labels is provided, the confidence score of the detection will be replaced with the label.
        :return: np.ndarray : The image with the bounding boxes drawn on it
        """
        font = cv2.FONT_HERSHEY_SIMPLEX
        for i, (xyxy, confidence, class_id, tracker_id) in enumerate(detections):
            color = (
                self.color.by_idx(class_id)
                if isinstance(self.color, ColorPalette)
                else self.color
            )
            text = (
                f"{confidence:0.2f}"
                if (labels is None or len(detections) != len(labels))
                else labels[i]
            )

            x1, y1, x2, y2 = xyxy.astype(int)
            text_width, text_height = cv2.getTextSize(
                text=text,
                fontFace=font,
                fontScale=self.text_scale,
                thickness=self.text_thickness,
            )[0]

            text_x = x1 + self.text_padding
            text_y = y1 - self.text_padding

            text_background_x1 = x1
            text_background_y1 = y1 - 2 * self.text_padding - text_height

            text_background_x2 = x1 + 2 * self.text_padding + text_width
            text_background_y2 = y1

            cv2.rectangle(
                img=frame,
                pt1=(x1, y1),
                pt2=(x2, y2),
                color=color.as_bgr(),
                thickness=self.thickness,
            )


        return frame
```

```python
import cv2
import os

# define paths
SOURCE_DIR = "BirdsScrap/BIRD/"
DEST_DIR = "BirdsScrapCrop/"


# # create instance of BoxAnnotator
box_annotator = BoxAnnotator(color=ColorPalette(), thickness=3, text_thickness=3, text_scale=1)

# loop through each species folder
for species_folder in os.listdir(SOURCE_DIR):

    species_path = os.path.join(SOURCE_DIR, species_folder)

    # create corresponding cropped species folder
    cropped_species_path = os.path.join(DEST_DIR, species_folder+"_crop")
    os.makedirs(cropped_species_path, exist_ok=True)

    # loop through each image in the species folder
    for image_filename in os.listdir(species_path):
        image_path = os.path.join(species_path, image_filename)
        print(image_path)
        # load image
        if image_path.split(".")[-1]=='svg':
          continue

        image = cv2.imread(image_path)

        # model prediction on single frame and conversion to supervision Detections
        try:
          results = model(image)
        except AssertionError:
          print("Skipping frame due to AssertionError.")
          continue

        detections = Detections(
            xyxy=results[0].boxes.xyxy.cpu().numpy(),
            confidence=results[0].boxes.conf.cpu().numpy(),
            class_id=results[0].boxes.cls.cpu().numpy().astype(int)
        )
        if len(results[0])==0:
          continue

        # loop through each bird detection
        for i in results[0]:
            # crop bird image
            bird_image = image[int(i.boxes.xyxy.cpu().numpy()[0][1]):int(i.boxes.xyxy.cpu().numpy()[0][3]),
                                int(i.boxes.xyxy.cpu().numpy()[0][0]):int(i.boxes.xyxy.cpu().numpy()[0][2])]

            black_canvas = np.zeros((300, 300, 3), dtype=np.uint8)

            # calculate the position to place the cropped image on the black canvas
            x_offset = (300 - bird_image.shape[1]) // 2
            y_offset = (300 - bird_image.shape[0]) // 2
            if bird_image.shape[0] < 100 or bird_image.shape[1] < 100:
              continue
            # resize the bird image if necessary
            if bird_image.shape[0] > 300 or bird_image.shape[1] > 300:
                resized_image = cv2.resize(bird_image, (300, 300))
            else:
                resized_image = bird_image

            # calculate the updated offsets for the resized image
            x_offset = (300 - resized_image.shape[1]) // 2
            y_offset = (300 - resized_image.shape[0]) // 2

            # paste the resized image onto the black canvas
            black_canvas[y_offset:y_offset + resized_image.shape[0], x_offset:x_offset + resized_image.shape[1]] = resized_image


            # save image with black borders to cropped species folder
            cropped_image_filename = os.path.splitext(image_filename)[0] + "_" + str(i.boxes.cls.cpu().numpy()[0]) + ".jpg"
            cropped_image_path = os.path.join(cropped_species_path, cropped_image_filename)

            # resize the image to 300x300 pixels
            resized_image = cv2.resize(black_canvas, (300, 300))

            cv2.imwrite(cropped_image_path, resized_image)

            # save cropped bird image to cropped species folder
            # cropped_image_filename = os.path.splitext(image_filename)[0] + "_" + str(i.boxes.cls.cpu().numpy()[0]) + ".jpg"
            # cropped_image_path = os.path.join(cropped_species_path, cropped_image_filename)
            # cv2.imwrite(cropped_image_path, bird_image)
```
